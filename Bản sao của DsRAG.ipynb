```python
from dsrag.knowledge_base import KnowledgeBase
from dsrag.llm import OpenAIChatAPI
from dsrag.reranker import NoReranker

llm = OpenAIChatAPI(model='gpt-4o-mini')
reranker = NoReranker()

kb = KnowledgeBase(kb_id="new_papers", reranker=reranker, auto_context_model=llm)

search_queries = ["KẾT QUẢ NGHIÊN CỨU PHẪU THUẬT ĐIỀU TRỊ TỔN THƯƠNG BÚP NGÓN TAY ĐỨT RỜI BẰNG PHƯƠNG PHÁP GHÉP BÚP là gì?"]
```

```python
kb.add_document(
    doc_id="BÀI BÁO MỚI",
    file_path=file_path,
    document_title="Báo cáo y khoa mới",
    file_parsing_config={
        "use_vlm": True,
        "vlm_config": {
            "provider": "gemini",
            "model": "gemini-2.0-flash",
        }
    }
)
```

```python
search_queries = ["KẾT QUẢ NGHIÊN CỨU PHẪU THUẬT ĐIỀU TRỊ TỔN THƯƠNG BÚP NGÓN TAY ĐỨT RỜI BẰNG PHƯƠNG PHÁP GHÉP BÚP là gì?"]
results = kb.query(search_queries)
for segment in results:
    print(segment['content'])
    print("#######------------------------------#######\n")
```
